### Now, with dropped columns, need to remove plants with missing relevant data
- dropNA IF missing MW, Capacity Factor, 

```python
#drop row if missing information in specific column
# save number of rows before dropping
numRows = []
for df in df_arr:
    numRows.append(df.shape[0])
numRows
```

```python
subsetCols = ['MW','Capacity_Factor','CO2_tons/MWh','Heat_rate_MMBtu/MWh','Nox_lbs/MWh']
```

```python
numRowsAfter = []
for df in df_arr:
    df.dropna(subset = subsetCols, axis = 0, inplace = True)
    numRowsAfter.append(df.shape[0])
numRowsAfter
```

```python
numRowsRemoved = []
for i in range(len(numRows)):
    numRowsRemoved.append(numRows[i]-numRowsAfter[i])

numRowsRemoved
```

```python
#Total # of peaker plants analyzed 
print(sum(numRowsAfter))
```
